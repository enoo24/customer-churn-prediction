```python
# import the packages
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score
```

<p>We are using F1 Score as model performance metric because its more appropriate in an imbalanced class problem</p>

<h3>Get the data</h3>

```python
train = pd.read_csv('dataset/train.csv')
train.head()
```

```python
test = pd.read_csv('dataset/test.csv')
test.head()
```

<p>We will use the test set only for final report on model performance</p>

```python
# scale the features and split the train data into train-dev set
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = train.drop(['churn'], axis=1)
y = train['churn']

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

X_train, X_dev, y_train, y_dev = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

<h3>Building neural network model</h3>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, BatchNormalization, Dropout
from tensorflow.keras import optimizers
```

```python
tf.config.experimental.list_physical_devices('GPU')
```

```python
def build_model1_three_hidden_layers():
    model = Sequential()
    model.add(Dense(32, input_shape=(X_train.shape[1],)))
    model.add(Dense(32, Activation('relu')))
    model.add(Dense(64, Activation('relu')))
    model.add(Dense(128, Activation('relu')))
    model.add(Dense(1))
    
    learning_rate = 0.001
    optimizer = optimizers.Adam(learning_rate)
    model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                 optimizer=optimizer,
                 metrics=['accuracy'])
    return model

model1 = build_model1_three_hidden_layers()
model1.summary()
```

```python
EPOCHS = 100
batch_size = 16

with tf.device('/GPU:0'):
    history = model1.fit(X_train,
                        y_train,
                        batch_size=batch_size,
                        epochs=EPOCHS,
                        verbose=1,
                        shuffle=True,
                        steps_per_epoch=int(X_train.shape[0] / batch_size),
                        validation_data=(X_dev, y_dev))
```

```python
model1_hist = pd.DataFrame(history.history)
model1_hist['epoch'] = history.epoch
model1_hist
```

```python
import matplotlib.pyplot as plt

plt.plot(model1_hist['accuracy'])
plt.plot(model1_hist['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Cross-validation'], loc='best')
plt.show()
```

```python
plt.plot(model1_hist['loss'])
plt.plot(model1_hist['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Cross-validation'], loc='best')
plt.show()
```

```python
loss, accuracy = model1.evaluate(X_dev, y_dev, verbose=2)
print('Accuracy: {}'.format(accuracy))
```

```python
import seaborn as sns

y_pred = (model1.predict(X_dev) >= 0.5).reshape(-1)

print('F1 score: {:.3f}'.format(f1_score(y_dev, y_pred)))
conf_matrix = pd.crosstab(y_dev, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(conf_matrix, annot=True, fmt='g');
```

<p>It's overfitting the training set. Let's try to add a regularization term</p>

```python
from keras.regularizers import l2

def build_model2_three_hidden_layers_with_regularization():
    model = Sequential()
    model.add(Dense(32, input_shape=(X_train.shape[1],), kernel_regularizer=l2(0.01)))
    model.add(Dense(32, Activation('relu'), kernel_regularizer=l2(0.01)))
    model.add(Dense(64, Activation('relu'), kernel_regularizer=l2(0.01)))
    model.add(Dense(128, Activation('relu'), kernel_regularizer=l2(0.01)))
    model.add(Dense(1))
    
    learning_rate = 0.001
    optimizer = optimizers.Adam(learning_rate)
    model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                 optimizer=optimizer,
                 metrics=['accuracy'])
    return model

model2 = build_model2_three_hidden_layers_with_regularization()
model2.summary()
```

```python
EPOCHS = 100
batch_size = 16

with tf.device('/GPU:0'):
    history = model2.fit(X_train,
                        y_train,
                        batch_size=batch_size,
                        epochs=EPOCHS,
                        verbose=1,
                        shuffle=True,
                        steps_per_epoch=int(X_train.shape[0] / batch_size),
                        validation_data=(X_dev, y_dev))
```

```python
model2_hist = pd.DataFrame(history.history)
model2_hist['epoch'] = history.epoch
model2_hist
```

```python
plt.plot(model2_hist['accuracy'])
plt.plot(model2_hist['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Cross-validation'], loc='best')
plt.show()
```

```python
plt.plot(model2_hist['loss'])
plt.plot(model2_hist['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Cross-validation'], loc='best')
plt.show()
```

```python
loss, accuracy = model2.evaluate(X_dev, y_dev, verbose=2)
print('Accuracy: {}'.format(accuracy))
```

```python
y_pred = (model2.predict(X_dev) >= 0.5).reshape(-1)

print('F1 score: {:.3f}'.format(f1_score(y_dev, y_pred)))
conf_matrix = pd.crosstab(y_dev, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(conf_matrix, annot=True, fmt='g');
```

<h3>Neural network: combination of SMOTE over-sampling with random undersampling</h3>

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

over = SMOTE(sampling_strategy=0.2)
under = RandomUnderSampler(sampling_strategy=0.6)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)

X_train_res, y_train_res = pipeline.fit_resample(X_train, y_train)

EPOCHS = 100
batch_size = 16

with tf.device('/GPU:0'):
    history = model2.fit(X_train_res,
                        y_train_res,
                        batch_size=batch_size,
                        epochs=EPOCHS,
                        verbose=1,
                        shuffle=True,
                        steps_per_epoch=int(X_train_res.shape[0] / batch_size),
                        validation_data=(X_dev, y_dev))
```

```python
model2_with_smote_and_under_hist = pd.DataFrame(history.history)
model2_with_smote_and_under_hist['epoch'] = history.epoch
model2_with_smote_and_under_hist
```

```python
plt.plot(model2_with_smote_and_under_hist['accuracy'])
plt.plot(model2_with_smote_and_under_hist['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Cross-validation'], loc='best')
plt.show()
```

```python
plt.plot(model2_with_smote_and_under_hist['loss'])
plt.plot(model2_with_smote_and_under_hist['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Cross-validation'], loc='best')
plt.show()
```

```python
loss, accuracy = model2.evaluate(X_dev, y_dev, verbose=2)
print('Accuracy: {}'.format(accuracy))
```

```python
y_pred = (model2.predict(X_dev) >= 0.5).reshape(-1)

print('F1 score: {:.3f}'.format(f1_score(y_dev, y_pred)))
conf_matrix = pd.crosstab(y_dev, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(conf_matrix, annot=True, fmt='g');
```

<p>It seems like tuned model2 with regularization method generates a higher F1 Score than any other methods</p>

<h3>Put all the pieces together</h3>

```python
X_test = test.drop(['churn'], axis=1)
y_test = test['churn']

X_test_scaled = scaler.transform(X_test)

EPOCHS = 100
batch_size = 16

with tf.device('/GPU:0'):
    history = model2.fit(X_scaled,
                        y,
                        batch_size=batch_size,
                        epochs=EPOCHS,
                        verbose=1,
                        shuffle=True,
                        steps_per_epoch=int(X_scaled.shape[0] / batch_size),
                        validation_data=(X_test_scaled, y_test))
```

```python
y_pred = (model2.predict(X_test_scaled) >= 0.5).reshape(-1)

print('F1 score: {:.3f}'.format(f1_score(y_test, y_pred)))
conf_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(conf_matrix, annot=True, fmt='g');
```
